## Estimating homographies

```python
# import the necessary packages
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import glob
import random
```

```python
## load the images 
path = glob.glob("images/*.tif")
img_list = []

for img in path:
    image = cv.imread(img)
    """ smoothed = cv.GaussianBlur(image, (9, 9), 10) 
    unsharped = cv.addWeighted(image, 2, smoothed, -1, 0) """
    image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    img_list.append(image)

    if not cv.findChessboardCornersSB(image, (13,12), None)[0]:
        print("No corners found", img)
```

```python
# display one random image
## display some random images from the dataset
fig = plt.figure(figsize=(15, 15))

img = img_list[9]
plt.imshow(img, cmap='gray')
plt.show()

print(f"{len(img_list)} images loaded")
print(f"Image shape: {img.shape}")
```
